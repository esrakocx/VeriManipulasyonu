```python
# ordinal tanımlama

from pandas.api.types import CategoricalDtype
```

```python
df["cut"].head()
```

```python
df.cut = df["cut"].astype(CategoricalDtype(ordered = True))
```

```python
df.cut.head() #ordinal (sıralı) hale geldi ama yanlış sırada
```

```python
cut_kategoriler = ["Fair", "Good", "Very Good", "Premium", "Ideal"]
```

```python
df.cut = df.cut.astype(CategoricalDtype(categories = cut_kategoriler, ordered = True))
```

```python
df.cut.head(1)
```

```python
# barplot
```

```python
df["cut"].value_counts().plot.barh()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.plot(x, y)
plt.xlabel("X ekseni")
plt.ylabel("Y ekseni")
plt.title("Matplotlib ile basit bir çizgi grafiği")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'X': [1, 2, 3, 4],
    'Y': [10, 20, 25, 30],
    'Kategori': ['A', 'A', 'B', 'B']
})
sns.lineplot(x='X', y='Y', hue='Kategori', data=data)
sns.set_theme()
plt.title('Seaborn ile çizgi grafiği')
plt.show()
```
